### Описание данных

Анализируемые данные состоят из 2 датасетов с заведениями общественного питания Москвы. Датасеты составлены на основе данных сервисов Яндекс Карты и Яндекс Бизнес на лето 2022 года.

Файл `/datasets/rest_info.csv` содержит информацию о заведениях:
- `name` — название заведения;
- `address` — адрес заведения;
- `district` — административный район, в котором находится заведение, например Центральный административный округ;
- `category` — категория заведения, например «кафе», «пиццерия» или «кофейня»;
- `hours` — информация о днях и часах работы;
- `rating` — рейтинг заведения по оценкам пользователей в Яндекс Картах (высшая оценка — 5.0);
- `chain` — число, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки):
    - `0` — заведение не является сетевым;
    - `1` — заведение является сетевым.
- `seats` — количество посадочных мест.

Файл `/datasets/rest_price.csv` содержит информацию о среднем чеке в заведениях:
- `price` — категория цен в заведении, например «средние», «ниже среднего», «выше среднего» и так далее;
- `avg_bill` — строка, которая хранит среднюю стоимость заказа в виде диапазона;
- `middle_avg_bill` — число с оценкой среднего чека, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Средний счёт»:
Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт.
- `middle_coffee_cup` — число с оценкой одной чашки капучино, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Цена одной чашки капучино» по аналогии с `middle_avg_bill`.

## 2. Предобработка данных

Подготовим данные к исследовательскому анализу. Для начала скорректируем размерность данных в столбце `chain`.

```python
df['chain'] = pd.to_numeric(df['chain'], downcast='integer')
```

Проверим результат оптимизации.

```python
df['chain'].dtypes
```

Изучим пропущенные значения в данных: посчитаем их количество и долю в каждом столбце датафрейме.

```python
missing_values = pd.DataFrame({
    'number_of_missing_values': df.isnull().sum(),
    'percentage_of_missing_values(%)': (df.isnull().mean() * 100).round(0).astype(int)
})

missing_values.sort_values(by='percentage_of_missing_values(%)', ascending=False)
```

Данные о количестве посадочных мест, среднем чеке и часах работы обычно добавляются на карты представителями заведений. Категория цен в заведении предположительно может присваиваться автоматически на основе среднего чека. 

- Доля пропусков в столбце `hours` относительно небольшая. Возможно, этот пункт не успели или забыли заполнить. Тогда пропуски будут относиться к типу MCAR, чтобы удаление строк с ними не влияло на результат анализа — оставим без изменений.
- Доли пропусков в столбцах `seats`, `avg_bill`, `price` достаточно велики, чтобы считать их случайными. На данном этапе сложно определить факторы, которые влияют на эти значения. Отнесем пропуски к типу MNAR и оставим без изменений.

На пропуски в столбцах `middle_avg_bill` и `middle_coffee_cup` влияют значения столбца `avg_bill`, значит они относятся к типу MAR. Можно заполнить пропуски на основе медианы или среднего, сгруппировав данные по колонкам `price`, `category`, `district`. Из-за высокой доли пропущенных значений такое заполнение может исказить результаты анализа.

Дополнительно проверим категориальные и дискретные данные на скрытые пропуски с помощью вывода уникальных значений.

```python
for column in [2, 4, 6, 7, 8, 9]:
    print(f'Уникальные значения в столбце "{df.columns[column]}":')
    print(df.iloc[::, column].sort_values().unique()) 
    print()
    
```

Значения-заглушки в столбцах `category`, `district`, `rating`, `chain`, `seats`, `price` не обнаружены.

Проверим датафрейм на явные и неявные дубликаты, предварительно нормализовав данные в текстовых столбцах.

```python
# Перевод данных в нижних регистр с удалением лишних пробелов и сохранение нормализованных значений в новых столбцах
df = df.assign(
    low_name = df['name'].str.strip().str.lower(),
    low_address = df['address'].str.strip().str.lower(),
    low_hours = df['hours'].str.strip().str.lower(),
    low_avg_bill = df['avg_bill'].str.strip().str.lower())
```

Сравним количество уникальных значений в столбцах до и после нормализации.

```python
columns = ['name', 'address', 'hours', 'avg_bill']
low_columns = ['low_name', 'low_address', 'low_hours', 'low_avg_bill']
unique_values = pd.DataFrame({
    'original': [df[col].nunique() for col in columns],
    'low': [df[col].nunique() for col in low_columns]
}, index=columns)
unique_values['delta'] = unique_values['original'] - unique_values['low']

print(unique_values)
```

Нормализация привела к обнаружению 1 дубликата в столбце `address` и 102 дубликатов в столбце `name`, связанных с написанием в одного текста в разном регистре.

Расcчитаем количество явных дубликатов, то есть строк с совпадениями во всех столбцах.

```python
df.duplicated().sum()
```

Расcчитаем количество неявных дубликатов, то есть строк с совпадением в названии заведения и адресе.

```python
df.duplicated(subset=['low_name', 'low_address']).sum()
```

Удалим дублирующиеся строки.

```python
df.drop_duplicates(subset=['low_name', 'low_address'], inplace=True)
```

Очистим датафрейм от вспомогательных столбцов `low_hours`, `low_avg_bill`.

```python
df = df.drop(columns=['low_hours', 'low_avg_bill'])
```

Сохраним нормализованные данные в столбцах `name` и `address` с удалением low-версий этих столбцов. Приведем текст в формат с заглавными буквами, аналогичный первоначальному.

```python
df['name'] = df.pop('low_name').str.title()
df['address'] = df.pop('low_address').str.title()
```

Для дальнейшей работы создадим столбец `is_24_7` с обозначением того, что заведение работает ежедневно и круглосуточно, то есть 24/7:
  - логическое значение `True` — если заведение работает ежедневно и круглосуточно;
  - логическое значение `False` — в противоположном случае.

```python
df['is_24_7'] = (
    (df['hours'].str.contains('ежедневно', case=False, na=None)) & (df['hours'].str.contains('круглосуточно', case=False)) 
    )
```

Выведем 5 строк датафрейма после преобразования.

```python
df.head()
```

## 3. Исследовательский анализ данных

### Задача 1

Исследуем количество объектов общественного питания — групируем датафрейм по категориям, считаем количество адресов в каждой категории и их долю от всех заведений.

```python
grouped_categories = df.groupby('category')['address'].count().sort_values(ascending=False)

categories_stat = pd.DataFrame({
    'number_of_objects': grouped_categories,
    'objects_shere(%)': (grouped_categories / df.shape[0] * 100).round(0).astype(int)
    })

categories_stat
```

Визуализируем полученные значения на столбчатой диаграмме.

```python
grouped_categories.plot.bar(
    figsize=(8,4),
    xlabel='Категория',
    ylabel='Количество объектов',
    rot=45,
    title='Сравнение категорий заведений по кол-ву объектов',
    color='#60a57c')
plt.grid(axis='y')
```

Анализ показал преобладание кафе (28%), ресторанов (24%) и кофейен (17%) среди заведений Москвы.

### Задача 2

Исследуем распределение количества заведений по административным районам Москвы, а также отдельно распределение заведений каждой категории в Центральном административном округе Москвы.

```python
grouped_districts = df.groupby('district')['address'].count().sort_values(ascending=False)

districts_stat = pd.DataFrame({
    'number_of_objects': grouped_districts,
    'objects_shere(%)': (grouped_districts / df.shape[0] * 100).round(0).astype(int)
    })

districts_stat
```

Визуализируем полученные значения на столбчатой диаграмме.

```python
grouped_districts.plot.barh(
    figsize=(10,5),
    ylabel='Район',
    xlabel='Количество объектов',
    rot=0,
    title='Сравнение административных районов по кол-ву объектов',
    color='#60a57c')
plt.grid(axis='x')
```

Абсолютный лидер среди административных округов Москвы по количеству заведений общепита — Центральный АО (27%), который опережает другие районы в 2,5 раза. Исследуем распределение заведений каждой категории в ЦАО.

```python
grouped_central_categories = df[df['district'] == 'Центральный административный округ']\
                    .groupby('category')['address'].count().sort_values(ascending=False)

central_categories_stat = pd.DataFrame({
    'number_of_objects': grouped_central_categories,
    'objects_shere(%)': (grouped_central_categories / df[df['district'] == 'Центральный административный округ'].shape[0] * 100).round(0).astype(int)
})

central_categories_stat
```

```python
grouped_central_categories.plot.bar(
    figsize=(8,4),
    xlabel='Категория',
    ylabel='Количество объектов',
    rot=45,
    title='Сравнение категорий заведений по кол-ву объектов в ЦАО',
    color='#60a57c')
plt.grid(axis='y')
```

В Центральном административном округе большую долю заведений занимают рестораны — 30%, на втором и третьем месте — кафе (21%) и кофейни (19%).

### Задача 3

Изучим соотношение сетевых и несетевых заведений в целом по всем данным и в разрезе категорий заведения.

```python
grouped_chain = df.groupby(['chain'])['address'].count().sort_values(ascending=False)

chain_stat = pd.DataFrame({
    'number_of_objects': grouped_chain,
    'objects_shere(%)': (grouped_chain / df.shape[0] * 100).round(0).astype(int)
})

chain_stat
```

В Москве несетевых заведений (62%) значительно больше, чем сетевых (38%). Далее сравним количество сетевых и несетевых заведений в каждой категории.

```python
grouped_cat_chain = df.groupby(['category', 'chain'])['address'].count().unstack(fill_value=0)
grouped_cat_chain
```

```python
grouped_cat_chain.plot.bar(
    figsize=(8,4),
    xlabel='Категория',
    ylabel='Количество объектов',
    rot=45,
    title='Сравнение кол-ва сетевых и несетевых заведений',
    color=['#60a57c', '#8b6778'])
plt.grid(axis='y')
plt.legend(
    title='Тип заведения',
    labels=['0 - не сетевое', '1 - сетевое'])
```

Среди булочных, кофеен и пиццерий незначительно преобладают сетевые заведения. В остальных категориях больше несетевых заведений. Особенно сильный разрыв среди баров и кафе.

Сгруппируйте данные по названиям заведений и найдите топ-15 популярных сетей в Москве. Для них посчитайте значения среднего рейтинга. Под популярностью понимается количество заведений этой сети в регионе. К какой категории заведений они относятся? Результат сопроводите подходящими визуализациями.

```python
chains = df[df['chain']==1]

pop_chains = pd.DataFrame({
    'chains': chains.groupby('name')['address'].count(),
    'mean_rating': chains.groupby('name')['rating'].mean().round(2),
    'category': chains.groupby('name')['category'].apply(lambda x: x.mode()[0])
})
top_chains = pop_chains.sort_values(by='chains', ascending=False).head(15)
top_chains 
```

```python
top_chains['chains'].plot.barh(
    figsize=(10,5),
    ylabel='Название сети',
    xlabel='Количество объектов',
    rot=0,
    title='Сравнение сетей заведений по кол-ву объектов',
    color='#60a57c')
plt.grid(axis='x')
```

```python
mean_value = top_chains['mean_rating'].mean()

top_chains['mean_rating'].plot.barh(
    figsize=(10,5),
    ylabel='Название сети',
    xlabel='Средний рейтинг',
    rot=0,
    title='Сравнение сетей по среднему рейтингу',
    color='#60a57c')
plt.grid(axis='x'),
plt.xlim(3,5),
plt.axvline(mean_value, color="red", linestyle="--", linewidth=1.5, label=f"Средний рейтинг: {mean_value:.1f}"),
plt.legend() 
```
